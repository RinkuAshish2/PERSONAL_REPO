```python
!pip install transformers accelerate

import gradio as gr
import torch
from transformers import Blip2Processor, Blip2ForConditionalGeneration

# Load the BLIP-2 processor and model
model_name = "Salesforce/blip2-opt-2.7b"
processor = Blip2Processor.from_pretrained(model_name)  # Use Blip2Processor
model = Blip2ForConditionalGeneration.from_pretrained(
    model_name, torch_dtype=torch.float16, device_map="auto"
)

# ... (rest of the code remains the same)

# Function to generate captions
def generate_caption(image):
    # Preprocess image
    inputs = processor(images=image, return_tensors="pt").to(model.device)

    # Generate caption
    output = model.generate(**inputs)
    caption = processor.tokenizer.decode(output[0], skip_special_tokens=True)

    return caption

# Create a Gradio UI
interface = gr.Interface(
    fn=generate_caption,
    inputs=gr.Image(type="pil"),
    outputs="text",
    title="🖼➡📝 BLIP-2 Image Captioning",
    description="Upload an image, and the AI will generate a caption using Salesforce/blip2-opt-2.7b."
)

# Launch the Gradio app
interface.launch()
```
